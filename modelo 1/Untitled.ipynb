```python
#-------------------------------------------------------------------------------
#------------------DEFINE O MODELO----------------------------------------------
#-------------------------------------------------------------------------------
class Autoencoder(nn.Module):
    def __init__(self):
        # N, 50
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Linear(n,400),
            nn.Tanh(),
            nn.Linear(400,300),
            nn.Tanh(),
            nn.Linear(300,200),
            nn.Tanh(),
            nn.Linear(200,3),
            nn.Tanh(),
        )
        self.Decoder=nn.Sequential(
            nn.Linear(3,200),
            nn.Tanh(),
            nn.Linear(200,300),
            nn.Tanh(),
            nn.Linear(300,400),
            nn.Tanh(),
            nn.Linear(400,500),
            nn.Tanh(),
            nn.Linear(500,n),
        )
        self.project=nn.Linear(1,3)
        self.decoder=nn.Sequential(
            nn.Linear(6,100),
            nn.Tanh(),
            nn.Linear(100,200),
            nn.Tanh(),
            nn.Linear(200,450),
            nn.Tanh(),
            nn.Linear(450,50),
            nn.Tanh(),
            nn.Linear(50,1),
        )
    def forward(self, x, t,mode):
        if mode ==1:
            encoded = self.encoder(x)
            t=self.project(t)
            aux=torch.cat((encoded,t),1)
            decoded = self.decoder(aux)
        if mode == 2:
            encoded = self.encoder(x)
            decoded = self.Decoder(x)
        return decoded,encoded
```

```python
model = Autoencoder()
for m in model.modules():
    if isinstance(m, (nn.Conv2d, nn.Linear)):
        nn.init.orthogonal_(m.weight)
criterion = nn.MSELoss() #segundo a investigar
optimizer = torch.optim.Adam(model.parameters())#,lr=1e-4,weight_decay = 1e-5)
#optimizer = torch.optim.SGD(model.parameters(),lr=1e-4,weight_decay = 1e-5)#,momentum=0.5)
```
